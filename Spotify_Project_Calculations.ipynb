```python
#Cleaned dataset
url="https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv"
df = pd.read_csv(url)

df["track_album_release_date"] = pd.to_datetime(df["track_album_release_date"], errors="coerce")
df["release_year"] = df["track_album_release_date"].dt.year.astype("Int64")

df = df.drop(columns=["track_id","track_album_id","playlist_id","track_album_release_date"])
df = df[df["release_year"]>=2015]
df = df.fillna("Unknown")
```

```python
#calculate some basic statistics based on the genre first
df.groupby('playlist_genre').agg({
    'track_popularity' : ['mean', 'median'],
    'danceability' : ['mean', 'median']
})
```

```python
#We are going to want to create some graphs using the variables we are
#making the calculations from. Here's an example of something that might
#be helpful to display, the distribution of the track popularity for a
#specific genre, which would be the user inputted one
filtered_df = df[df['playlist_genre'] == 'edm']
sns.histplot(filtered_df['track_popularity'], bins=10, kde=True)
plt.show()
```

```python
#Here's a stacked histogram to display what we might want to show if we want
#more in depth description of a specific variable across all groups.
sns.histplot(data=df, x='track_popularity', hue='playlist_genre', multiple='stack', bins=8)
plt.show()
```

```python
#Here are just a few statistics we can generate for each of the subgenres.
#Having these calculations on hand will allow us to connect a specific average
#to an inputted subgenre.

#For example, the user says their preferred subgenre is classic rock.
#We can see that the mean danceability is 0.544.
#From there, we can look for songs that have around that danceability score to make a better recommendation!
df.groupby('playlist_subgenre').agg({
    'track_popularity' : ['mean', 'median'],
    'danceability' : ['mean', 'median']
})
#These are just a few variables and a few calculations, so we will want to do
#more than just this! This is just to help everyone get started.
```

```python
df.groupby('playlist_genre')[['energy', 'valence']].mean().sort_values('energy', ascending=False)

sns.scatterplot(data=df, x='energy', y='valence', hue='playlist_genre')
plt.title("Energy vs Valence by Genre")
plt.show()

# Energy vs Valence
# Goal: See how energetic songs relate to their mood (valence = positivity).
# Explanation: By grouping songs by genre and comparing their energy and valence scores, we can explore whether more energetic music also sounds happier.
# in this plot, genres like EDM and pop  cluster in higher energy and valence areas, suggesting these genres are both upbeat and lively.
```

```python
df[['tempo', 'danceability']].corr()
sns.scatterplot(data=df, x='tempo', y='danceability')
plt.title("Tempo vs Danceability")
plt.show()

# Tempo vs Danceability
# Goal: Check if faster songs are more danceable.
# Explanation: The scatterplot shows each song’s tempo (speed) and danceability.
# We can see that danceability doesn’t only depend on high tempo, many medium-tempo songs are still very danceable.
```

```python
# Calculate average danceability for each genre
avg_danceability = df.groupby('playlist_genre')['danceability'].mean().sort_values(ascending=False)

# Visualize
avg_danceability.plot(kind='bar', color='skyblue')
plt.title("Average Danceability by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Danceability")
plt.show()

# This calculation shows which genres tend to have songs that are easier to dance to.
# For example, if EDM or Latin appear at the top, it means those genres have songs with stronger rhythm and beat, making them more danceable overall.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select a few key numerical features
features = ['danceability', 'energy', 'valence', 'tempo', 'acousticness']

# Compute correlation matrix
corr = df[features].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Audio Features")
plt.show()

# This heatmap shows how different song features are connected. It helps us understand what makes songs sound energetic, happy, or calm.
# A positive correlation (red) means features increase together (e.g., energy and valence).
# A negative correlation (blue) means one increases while the other decreases (e.g., acousticness and energy).
```

```python
avg_duration = df.groupby('playlist_genre')['duration_ms'].mean() / 60000  # convert ms to minutes
avg_duration.plot(kind='bar', color='skyblue')
plt.title("Average Song Duration by Genre")
plt.ylabel("Duration (minutes)")
plt.show()
```

```python
sns.scatterplot(data=df, x='loudness', y='energy', hue='playlist_genre')
plt.title("Energy vs Loudness by Genre")
plt.xlabel("Loudness (dB)")
plt.ylabel("Energy")
plt.show()

# Explanation:
# Louder songs often have higher energy. This visualization helps us see if that trend holds across genres.
# Genres like EDM or Rock may appear in higher loudness-energy zones, showing their strong and intense sound profile.
```

```python
sns.scatterplot(data=df, x='acousticness', y='valence', hue='playlist_genre')
plt.title("Valence vs Acousticness by Genre")
plt.xlabel("Acousticness")
plt.ylabel("Valence (Positivity)")
plt.show()

# Explanation:
# This helps us see if acoustic songs are associated with happier or calmer moods.
# If the correlation is low or negative, it suggests acoustic songs might not always sound "happy" (low valence).
```

```python
avg_features = df.groupby('playlist_genre')[['energy', 'danceability']].mean().sort_values('energy', ascending=False)
avg_features.plot(kind='bar', figsize=(8,5))
plt.title("Average Energy and Danceability by Genre")
plt.ylabel("Average Value")
plt.show()

# Explanation:
# This comparison shows which genres are both high-energy and danceable.
# It helps us identify which genres are best suited for upbeat or workout-style music recommendations.
```

```python
avg_popularity = df.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
avg_popularity.plot(kind='bar', color='orange')
plt.title("Average Track Popularity by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.show()
```

```python
corr_value = df['energy'].corr(df['track_popularity'])
print(f"Correlation between Energy and Popularity: {corr_value:.2f}")
```

```python
avg_duration = df.groupby('playlist_genre')['duration_ms'].mean() / 60000  # convert ms to minutes
plt.figure(figsize=(8,5))
avg_duration.plot(kind='bar', color='lightgreen')
plt.title("Average Song Duration by Genre (minutes)")
plt.xlabel("Genre")
plt.ylabel("Duration (minutes)")
plt.show()
```

```python
# Group data by genre
avg_features = df.groupby('playlist_genre')[['track_popularity', 'energy']].mean().sort_values('track_popularity', ascending=False)

avg_features.plot(kind='bar', figsize=(8,5), color=['#1DB954', '#ff7f0e'])
plt.title("Average Popularity and Energy by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Value")
plt.legend(["Popularity", "Energy"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_popularity = df.groupby('playlist_genre')['track_popularity'].mean().sort_values()

plt.figure(figsize=(8,5))
bars = plt.barh(avg_popularity.index, avg_popularity.values, color=plt.cm.plasma(avg_popularity.values / max(avg_popularity.values)))
plt.title("Average Popularity by Genre (Spotify Dataset)")
plt.xlabel("Average Popularity")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(7,5))
sns.regplot(x='energy', y='track_popularity', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Relationship Between Energy and Popularity")
plt.xlabel("Energy")
plt.ylabel("Popularity")
plt.show()
```
